```python
features = ['Batting Avg.','Strike rate','Wicket keeping','economy','avg speed','fitness index','Sixes','fours','Fast bowling','Leg Spin','Off Spin','against fast bowl','against leg spin','against off spin','Fanbase','Catch percentage']
```

```python
X = df[features]
```

```python
X.describe()
```

```python
X.head()
```

```python
Y = df['Not selected']
```

```python
Y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
train_X , val_X , train_Y , val_Y = train_test_split(X,Y,random_state = 0)
```

```python
#Decision tree
from sklearn import tree
from sklearn.metrics import mean_absolute_error
model1 = tree.DecisionTreeClassifier(random_state = 1)
model1 = model1.fit(train_X,train_Y)
val_predictions1 = model1.predict(val_X)
print(mean_absolute_error(val_Y , val_predictions1))
```

```python
#Random forest classifier
from sklearn.ensemble import RandomForestClassifier
model2 = RandomForestClassifier(random_state = 1)
model2 = model2.fit(train_X,train_Y)
val_predictions2 = model2.predict(val_X)
print(mean_absolute_error(val_Y , val_predictions2))
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
model3 = LogisticRegression(random_state = 1)
model3 = model3.fit(X,Y)
```

```python
#svm
from sklearn import svm
model4 = svm.SVC(random_state = 1)
model4 = model4.fit(X,Y)
```

```python
#knn
from sklearn.neighbors import KNeighborsClassifier
model5 = KNeighborsClassifier()
model5 = model5.fit(train_X,train_Y)
val_predictions5 = model5.predict(val_X)
print(mean_absolute_error(val_Y , val_predictions5))
```

```python
#sklearn.ensemble.GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingClassifier
model6 = GradientBoostingClassifier()
model6 = model6.fit(train_X,train_Y)
val_predictions6 = model6.predict(val_X)
print(mean_absolute_error(val_Y , val_predictions6))
```

```python
#naive bayes
from sklearn.naive_bayes import GaussianNB
model7 = GaussianNB()
model7 = model7.fit(train_X,train_Y)
val_predictions7 = model6.predict(val_X)
print(mean_absolute_error(val_Y , val_predictions7))
```
